```python
from gliner import GLiNER

# Load the pre-trained GLiNER model
model = GLiNER.from_pretrained("knowledgator/gliner-multitask-large-v0.5")

# Input text (patient case or medical summary)
text = """
The patient is experiencing chronic pain in the lower back and has been diagnosed with osteoarthritis.
They also report having a history of hypertension and diabetes.
"""

# Define the true entities you want to check (the actual diseases and symptoms)
true_entities = ["hypertension", "chronic pain"]

# Use the model to predict the entities
entities = model.predict_entities(text, ["disease", "symptom"])

# Print the extracted entities with their labels, but only if the entity text matches the true entities
for entity in entities:
    if entity["text"].lower() in map(str.lower, true_entities):  # Case-insensitive check
        print(entity["text"], "=>", entity["label"])
```

```python
# Initialize counters
predicted_count = 0
true_count = 0

# Iterate over the predicted entities
for entity in entities:
    predicted_count += 1  # Count all predicted entities
    if entity["text"].lower() in map(str.lower, true_entities):  # Case-insensitive check
        true_count += 1  # Count true entities

# Display the counts
print(f"Total predicted entities: {predicted_count}")
print(f"Total true entities found: {true_count}")
```

```python
# Initialize counters
predicted_count = 0
true_count = 0
true_positive = 0
false_positive = 0
false_negative = 0

# Create a set of predicted entities (lowercased for case-insensitive comparison)
predicted_texts = set([entity["text"].lower() for entity in entities])

# Iterate over the true entities to check for false negatives
for true_entity in true_entities:
    if true_entity.lower() in predicted_texts:
        true_positive += 1  # It's a true positive
    else:
        false_negative += 1  # It's a false negative

# Iterate over the predicted entities to check for false positives
for entity in entities:
    predicted_count += 1  # Count all predicted entities
    if entity["text"].lower() not in map(str.lower, true_entities):  # False positive check
        false_positive += 1  # It's a false positive

# Display the counts
print(f"Total predicted entities: {predicted_count}")
print(f"True positives: {true_positive}")
print(f"False positives: {false_positive}")
print(f"False negatives: {false_negative}")
```

```python
# Calculate Precision, Recall, F1 Score, and Accuracy
precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
accuracy = (true_positive) / (predicted_count) if predicted_count > 0 else 0
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import mlflow
import mlflow.sklearn
from gliner import GLiNER

# Start a new MLflow run
with mlflow.start_run():

    # Load the pre-trained GLiNER model
    model = GLiNER.from_pretrained("knowledgator/gliner-multitask-large-v0.5")

    # Input text (patient case or medical summary)
    text = """
    The patient is experiencing chronic pain in the lower back and has been diagnosed with osteoarthritis.
    They also report having a history of hypertension and diabetes.
    """

    # Define the true entities you want to check (the actual diseases and symptoms)
    true_entities = ["hypertension", "chronic pain"]

    # Use the model to predict the entities
    entities = model.predict_entities(text, ["disease", "symptom"])

    # Initialize counters
    predicted_count = 0
    true_count = 0
    true_positive = 0
    false_positive = 0
    false_negative = 0
    true_negative = 0

    # Create a set of predicted entities (lowercased for case-insensitive comparison)
    predicted_texts = set([entity["text"].lower() for entity in entities])

    # Iterate over the true entities to check for false negatives
    for true_entity in true_entities:
        if true_entity.lower() in predicted_texts:
            true_positive += 1  # It's a true positive
        else:
            false_negative += 1  # It's a false negative

    # Iterate over the predicted entities to check for false positives
    for entity in entities:
        predicted_count += 1  # Count all predicted entities
        if entity["text"].lower() not in map(str.lower, true_entities):  # False positive check
            false_positive += 1  # It's a false positive


    # Calculate Precision, Recall, F1 Score, and Accuracy
    precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
    recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Log metrics to MLflow
    mlflow.log_metric("True positives", true_positive)
    mlflow.log_metric("False positives", false_positive)
    mlflow.log_metric("False negatives", false_negative)
    mlflow.log_metric("Precision", precision)
    mlflow.log_metric("Recall", recall)
    mlflow.log_metric("F1 Score", f1_score)

    # Print the results (this will be displayed in MLflow's UI as well)
    print(f"Total predicted entities: {predicted_count}")
    print(f"True positives: {true_positive}")
    print(f"False positives: {false_positive}")
    print(f"False negatives: {false_negative}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1_score:.4f}")

    # Optionally, you can log other information like the model, parameters, or artifacts
    mlflow.log_param("Model", "knowledgator/gliner-multitask-large-v0.5")
    mlflow.log_param("Text", text)
```
